```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from matplotlib import style
import matplotlib
import h5py
import time
import json
import cv2
# import NaiveDE
# import SpatialDE
import scanpy as sc
import sklearn
import networkx as nx
import ot
%load_ext autoreload
%autoreload 2
# style.use('seaborn-dark')
style.use('seaborn-white')
```

```python
def read_DLPFC(gene_expression_file, location_file, image_file, scalefactor_json_file, spot_data, sample, min_count_per_spot=100,min_count_per_gene=100,remove_nan=True, fullres=False):
    adata = sc.read_10x_h5(gene_expression_file)
    adata.var_names_make_unique()
    print(adata.shape)

    # read in spatial locations
    all_locations = pd.read_csv(filepath_or_buffer=location_file, header=None, index_col=0, usecols=[0, 1, 4, 5])
    tissue_locations = all_locations[all_locations[1] == 1]
    spatial = []
    for spotname in adata.obs.index:
        spatial.append([tissue_locations.loc[spotname][5], tissue_locations.loc[spotname][4]])
    adata.obsm['spatial'] = np.array(spatial)

    # find pixel locations of each spot in the image, as well as spot size
    scalefactor_f = open(scalefactor_json_file)
    scalefactor = json.load(scalefactor_f)
    if fullres:
        spot_locations = adata.obsm['spatial']
        spot_radius = round(scalefactor['spot_diameter_fullres'] / 2.0)
    else:
        spot_locations = adata.obsm['spatial'] * scalefactor['tissue_hires_scalef']
        spot_radius = round(scalefactor['spot_diameter_fullres'] * scalefactor['tissue_hires_scalef'] / 2.0)
    scalefactor_f.close()

    # read in rgb values
    img = cv2.imread(image_file)
    rgb = []
    for i in range(adata.n_obs):
        x, y = round(spot_locations[i][0]), round(spot_locations[i][1])
        spot_mask = np.zeros((img.shape[0], img.shape[1]), np.uint8)
        cv2.circle(spot_mask, (x, y), radius=spot_radius, color=(255, 255, 255), thickness=-1)
        rgb.append(cv2.mean(img, spot_mask)[::-1][1:])
    adata.obsm['rgb'] = np.array(rgb)

    # read in colDataAll 
    adata.obs = spot_data[spot_data.sample_name==int(sample)][['tissue', 'row', 'col','imagerow','imagecol','sum_umi', 'sum_gene','subject', 'position', 'replicate', 'discard','cell_count','layer_guess', 'layer_guess_reordered', 'layer_guess_reordered_short']]
    adata.obs = adata.obs.rename(columns={'tissue':'in_tissue', 'row':'array_row', 'col':'array_col'})
    sc.pp.calculate_qc_metrics(adata, inplace=True)
    sc.pp.filter_genes(adata,min_counts=min_count_per_gene)
    sc.pp.filter_cells(adata,min_counts=min_count_per_spot)
    if remove_nan:
        adata = adata[~adata.obs['layer_guess_reordered'].isnull()]
    print(adata.shape)
    sc.pl.spatial(adata, color='layer_guess_reordered',spot_size=5)
    return adata

sample_list = ["151507", "151508", "151509","151510", "151669", "151670","151671", "151672", "151673","151674", "151675", "151676"]

spot_data = pd.read_csv('/n/fs/ragr-data/users/xinhao/DLPFC/colDataAll.csv',index_col=0,sep=',')
ge_file_name = '/n/fs/ragr-data/users/xinhao/DLPFC/{0}_filtered_feature_bc_matrix.h5'
location_file_name = '/n/fs/ragr-data/users/xinhao/DLPFC/{0}/tissue_positions_list.txt'
image_file_name = '/n/fs/ragr-data/users/xinhao/DLPFC/{0}/tissue_hires_image.png'
scalefactor_json_file_name = '/n/fs/ragr-data/users/xinhao/DLPFC/{0}/scalefactors_json.json'
adatas = {}

for sample in sample_list:
    print(sample,'-----')
    adatas[sample] = read_DLPFC(ge_file_name.format(sample), location_file_name.format(sample), image_file_name.format(sample), scalefactor_json_file_name.format(sample), spot_data, sample)

sample_groups = [["151507", "151508", "151509","151510"],[ "151669", "151670","151671", "151672"],[ "151673","151674", "151675", "151676"]]
layer_groups = [[adatas[sample_groups[j][i]] for i in range(len(sample_groups[j]))] for j in range(len(sample_groups))]
layer_to_color_map = {'Layer{0}'.format(i+1):sns.color_palette()[i] for i in range(6)}
layer_to_color_map['WM'] = sns.color_palette()[6]



```
